## Standard Library

**[time](https://docs.python.org/ko/3/library/time.html)** module 

**[random](https://docs.python.org/ko/3/library/random.html)** module 

```python
import time

# time.time
print(time.time())

# time.localtime
lt = time.localtime()

# time.asctime
print(time.asctime(lt))

# time.ctime
print(time.ctime())
```

```python
import random as rd # rd.xxx

# random.random
print(rd.random())

# random.randint
print(rd.randint(1,10))

# random.choice
foods = ['Apple', 'Melon', 'Rice', 'Pasta', 'Roll']
print(rd.choice(foods))

# random.shuffle
lst = [1,2,3,4,5]
rd.shuffle(lst)
print(lst)
```

## Matplotlib

- A Python library for data visualization
- [Matplotlib](https://matplotlib.org/stable/index.html)

```python
import matplotlib.pyplot as plt
```

```python
# plt.plot, plt.show
x = [1, 2, 3, 4]
y = [100, 200, 300, 150]

plt.plot(x, y)
plt.show()
```

```python
# plt.title, plt.xlabel, plt.ylabel

plt.plot(x,y)
plt.title('velocity per time')
plt.xlabel('time')
plt.ylabel('velocity')
plt.show()
```

```python
# plt.xlim, plt.ylim
plt.plot(x, y)
plt.title('velocity per time')
plt.xlabel('time')
plt.ylabel('velocity')

plt.xlim(1.5, 2.5)
plt.ylim(0, 1000)
# 자동으로 최적의 리미트 설정해줌 (설정 안했을 시에)

plt.show()
```

```python
# plt.xscale, plt.yscale
plt.plot(x, y)
plt.title('velocity per time')
plt.xlabel('time')
plt.ylabel('velocity')

plt.ylim(0,500)
plt.xscale('log') #linear 도 사용가능

plt.show()
```

```python
# plt.annotate
plt.plot(x, y)
plt.title('velocity per time')
plt.xlabel('time')
plt.ylabel('velocity')
plt.xlim(0,5)
plt.ylim(0,500)

plt.annotate('threshold', xy=(2,200), xytext=(2,300), arrowprops={'color':'blue'})
plt.show()
```

```python
# plt.legend

plt.plot([1,2,3], [100,300,500])
plt.plot([2,3,4], [200,300,400])
plt.plot([2,3,4], [300,350,100])

plt.title('velocity per time')
plt.xlabel('time')
plt.ylabel('velocity')

plt.legend(['car1','car2','car3'], loc = 'lower left') #loc = xxx
plt.show()
```

```python
# plt.subplot 여러 개의 그림을 한 곳에
plt.subplot(2,1,1)
plt.plot([1,2,3], [100,200,300])

plt.subplot(2,1,2)
plt.plot([1,2,3], [200,300,400])

plt.show()
```

```python
# plt.subplot
plt.subplot(2,2,1)
plt.plot([1,2,3], [100,300,500])
plt.title('1st')

plt.subplot(2,2,2)
plt.plot([2,3,4], [200,300,400])
plt.title('2nd')


plt.subplot(2,2,3)
plt.plot([2,3,4], [10,30,40])

plt.subplot(2,2,4)
plt.plot([2,3,4], [20,30,40])

plt.show()
```

```python
# line style = >  plt.plot([x], [y], '선')
# => - / -- / : / -.
plt.plot([1, 2, 3], [100, 300, 500], '-') 
plt.plot([2, 3, 4], [200, 300, 400], '--') 
plt.plot([1, 2, 3], [100, 150, 200], 'o')  # 동그라미 marker
plt.plot([1, 2, 3], [300, 400, 500], '-o') # 실선에 동그라미 marker

plt.title("velocity per time") 
plt.xlabel("time")
plt.ylabel("velocity")
plt.legend(["car1", "car2", "car3", "car4"], loc = 'lower right') 
plt.show()
```

```python
# line style 

plt.plot([1, 2, 3], [100, 300, 500], '-') 
plt.plot([2, 3, 4], [200, 300, 400], '-x') 
plt.plot([1, 2, 3], [100, 150, 200], '*') 
plt.plot([1, 2, 3], [300, 400, 500], '--D') 
plt.title("velocity per time") 
plt.xlabel("time")
plt.ylabel("velocity")
plt.legend(["car1", "car2", "car3", "car4"]) 
plt.show()
```

```python
# plt.bar 막대그래프
y = [3, 1, 3, 2, 3, 4]
x = range(1, len(y)+1) # range(1,7)
plt.bar(x, y, width=0.8, color="blue") 
plt.show()
```

```python
# plt.scatter 분포도
x = [170, 180, 160, 150, 166, 146, 177, 160, 184, 171]
y = [60, 77, 71, 45, 66, 44, 70, 66, 80, 70]
plt.scatter(x, y, marker ='P', c=x, s=y) #c는 컬러, s는 사이즈
plt.title("Weight per Height") 
plt.xlabel("height") 
plt.ylabel("weight") 
plt.show()
```

```python
# plt.pie
data = [1, 2, 4, 8] 
plt.pie(data)
plt.title("Pie Graph") 
plt.legend(["a", "b", "c", "d"]) 
plt.show()
```
